```python
import numpy as np

# c casos de test
# N , R, W , U y V

#open file
f = open("test1.txt", "w")
np.random.seed(0)

c = 100
f.write(str(c) + "\n")
n = 10
for i in range(c):
    n+= 10
    r = np.random.randint(1, 10000)
    if n == 1:
        w = 1
    else:
        w = np.random.randint(1, n)
    u = np.random.randint(1, 10)
    if u == 1:
        v = 1
    else:
        v = np.random.randint(1, u)
    f.write(str(n) + " " + str(r) + " " + str(w) + " " + str(u) + " " + str(v) + "\n")
    for j in range(n):
        a = np.random.randint(0, 1000)
        b = np.random.randint(0, 1000)
        f.write(str(a) + " " + str(b) + "\n")
f.close()
```

```python
# load file

# run test

# save results

# plot results


exp = 99

f = open("ej3_times.txt", "r")
times_d = []
times_nr = []
times_npc = []
times_dense = []
duration = 0
no_rank = 0
no_path_compression = 0
dense = 0
for i in range(exp):
    # grab the first 4 lines
    for j in range(4):
        #grab the line items separated by a space
        line = f.readline().split(" ")[0]
        if j == 0:
            duration = float(line)
            times_d.append(duration)
        elif j == 1:
            no_rank = float(line)
            times_nr.append(no_rank)
        elif j == 2:
            no_path_compression = float(line)
            times_npc.append(no_path_compression)
        elif j == 3:
            dense = float(line)
            times_dense.append(dense)
    

f.close()
```

```python
#plot times
import matplotlib.pyplot as plt
# add labels 
plt.xlabel('N')
plt.ylabel('Tiempo de ejecución')
# add title
plt.title('Comparación de tiempos de ejecución según método de resolución')

plt.plot(times_d, label = "Duración")
plt.plot(times_nr, label = "Sin optimizacion de union by rank")
plt.plot(times_npc, label = "Sin path compression")
plt.plot(times_dense, label = "Optimización para grafos densos")
plt.legend()
plt.show()

```
